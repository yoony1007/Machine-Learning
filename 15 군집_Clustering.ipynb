# 군집 (Clustering)
비지도 학습으로 비슷한 특성을 가지는 데이터들끼리 그룹으로 묶음
- 비지도 학습? Feature 값만 가지고 있음. Label 값이 없음
- 패턴을 찾아 비슷한 패턴 가지고 있는 X값끼리 묶어서 나눠주는 것
- 문제? 답이 없는 상태에서 학습했기 때문에 검증 결과의 정확도를 확인할 수 없음. 그룹으로 분류된 결과가 무엇인지 사람이 판독해야함

## 적용
- 고객 분류
    - 고객 데이터를 바탕으로 비슷한 특징의 고객들을 묶어 성향 파악
- 이상치 탐지
    - 어느 군집에도 묶이지 않는 데이터는 이상치일 가능성 높음
- 준지도학습
    - 레이블이 없는 데이터셋에 군집을 이용해 Label을 생성해 분류 지도학습 할 수 있음. 또는 레이블을 좀 더 세분화 할 수 있음


## k-mean (K-평균)
- 가장 널리 사용되는 군집 알고리즘 중 하나.
- 데이터셋을 K개의 군집으로 나눔. K는 하이퍼파라미터로 사용자가 지정
- 군집의 중심이 될 것 같은 임의의 지점(Centroid) 선택해 해당 중심에 가장 가까운 포인드들을 선택하는 기법

### 특징
- K-mean은 군집을 원 모양으로 간주
- 모든 특성은 동일한 Scale 가져야 함 
- 이상치에 취약

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib as mpl

mpl.rcParams['font.family'] = 'malgun ghtoic'
mpl.rcParams['axes.unicode_minus'] = True
```

```python
from sklearn.datasets import load_iris
data = load_iris()
X, y = data['data'], data['target']
```

```python
# K-mean - scaling 전처리 필요
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
from sklearn.cluster import KMeans

# 모델생성 - 몇개의 그룹으로 나눌지 지정
kmeans = KMeans(n_clusters = 3, random_state = 1)
# 학습
kmeans.fit(X_scaled)
```

```python
# 학습한 데이터셋의 각각의 원소들을 어떤 그룹으로 나눴는지 그 라벨 반환
cluster = kmeans.labels_
cluster
```

```python
X.shape, cluster.shape
```

```python
new_data = [
    [0.8, 0.9, 1.5, 3.7],
    [3, 4, 7, 1]
]
kmeans.predict(new_data)
```

```python
df = pd.DataFrame(X, columns = data['feature_names'])
df['cluster'] = cluster
df['y'] = y
df.head(10)
```

```python
df['cluster'].value_counts()
```

```python
df.groupby(['y', 'cluster'])['sepal length (cm)'].count()
```

## Inertia value를 이용한 적정 군집수 판단
- inertia 
    - 군집내 데이터들과 중심간의 거리의 합. 군집의 응집도 나타내는 값
    - 값이 작을수록 응집도가 높게 군집화가 잘되었다고 평가    
    - KMean의 inertia_ 속성으로 조회
    - 군집 단위별로 inertia 값 조회한 후, 급격히 떨어지는 지점이 적정 군집수라 판단

```python
# inertia 조회
kmeans.inertia_
```

```python
# 군집 수의 후보
k_list = range(2, 6)
# inertia 점수저장할 리스트
inertia_list = []

for k in k_list:
    model = KMeans(n_clusters = k, random_state = 1)
    model.fit(X_scaled)
    inertia_list.append(model.inertia_)
```

```python
inertia_list
```

```python
# 그래프
plt.figure(figsize = (7,6))
plt.plot(k_list, inertia_list)
plt.xticks(k_list)
plt.show()
```

### 평가 : 실루엣 지표

- 실루엣 계수 (silhouette coefficient)
    - 개별 관측치가 해당 군집 내의 데이터와 얼마나 가깝고 가장 가까운 다른 군집과 얼마나 먼지를 나타내는 지표
    - b-a / max(a,b) - a: 해당 군집 내 원소과의 거리 b: 가장 가까운 다른 군집 원소와의 거리
    - -1 ~ 1 사이의 값이 나옴
    - 1에 가까울 수록 좋은 지표
    - 0에 가까울수록 다른 군집과 가까움 나타냄 -> 어느 군집에 속해도 큰 차이 없는 것
- silhouette_samples()
    - 개별 관측치의 실루엣 계수 반환
- silhouette_score()
    - 실루엣 계수들의 평균
- 좋은 군집화의 지표
    - 실루엣 계수 평균이 1에 가까울수록 좋음
    - 실루엣 계수 평균과 개별 군집의 실루엣 계수 평균의 편차가 크지 않아야 함

```python
from sklearn.metrics import silhouette_score, silhouette_samples
```

```python
# X, cluster
s_coef = silhouette_samples(X_scaled, cluster)
s_coef
```

```python
s_coef.shape
```

```python
silhouette_score(X_scaled, cluster)
```

```python
np.mean(s_coef)
```
